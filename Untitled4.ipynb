```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris #for linear regression, loading iris dataset
```

```python
data=load_iris()
data.feature_names#featfures of iris_data set
```

```python
data.target_names#flower species
```

```python
data.target
```

```python
x=data.data #data : value of independent colmns 
x.shape #check shape
```

```python
y=data.target
y.shape
```

```python
# beoz sklearn dont like 1D array or vectoers we are going to reshape it
y=y.reshape(-1,1)
y.shape
```

```python
plt.figure(figsize=(18,8),dpi=100)
plt.scatter(x.T[0],x.T[2])
plt.title('IRIS Petal ans sepal length', fontsize=20)
plt.ylabel('petal length')
plt.xlabel('sepal length')
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
x_train,x_test,y_train,y_test=train_test_split(x,y, test_size=0.20)
```

```python
lr =LinearRegression()
iris_model =lr.fit(x_train, y_train)
prediction = iris_model.predict(x_test)
```

```python
prediction = np.asarray(prediction).flatten()
yerr = np.abs(y_test - prediction)

plt.errorbar(range(1, len(y_test) + 1), y_test, yerr=yerr, fmt='^k', ecolor='red')
plt.show()
```
